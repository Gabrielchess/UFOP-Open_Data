```python
import regex as re
import pandas as pd
import os
import PyPDF2

from tabula.io import read_pdf
from PyPDF2 import PdfReader
```

```python
path = 'C:/Users/James Bond/Desktop/DataAnaliseUFOP/MatrizCurricular'
cursos = os.listdir(path)
```

```python
#Ler matriz currucular
dfs = []

os.chdir(path)

for i in range(len(cursos)):
    tables = read_pdf(cursos[i],pages="all")

    # creating an object and reader
    file = open(cursos[i], 'rb')
    fileReader = PyPDF2.PdfFileReader(file)
    
    #Ler arquivo
    text = fileReader.pages[0].extract_text()

    #Identificar Curso e Modalidade
    curso = re.findall(r'(?=CURSO DE).+?(?= \()', text)[0]
    modalidade = re.findall(r'(?=\().+?(?=\) )', text)[0][1:]
    
    for j in range (len(tables)):
        tables[j]['curso'] = curso
        tables[j]['modalidade'] = modalidade
        
    dfs.append(tables)
```

```python
obrigatorias = []

for i in range (len(dfs)-1):
    ob = dfs[i][0].iloc[1:].dropna(axis=1, how='all')
    ob.rename(columns={ob.columns[0]: 'CÓDIGO', ob.columns[1]: 'DISCIPLINAS OBRIGATÓRIAS'},inplace=True)                       
    obrigatorias.append(ob)
```

```python
foo = pd.concat(obrigatorias)
foo[foo['CÓDIGO'].isna() == True]
#foo['Unnamed: 1'].value_counts()
```
